```python
from pathlib import Path

import automol
import elstruct
import numpy as np
```

```python
saddle_bc_out_str = Path("saddle_bc.out").read_text()
well_c_out_str = Path("well_c.out").read_text()

saddle_bc_geo = elstruct.reader.opt_geometry("gaussian16", saddle_bc_out_str)
well_c_geo = elstruct.reader.opt_geometry("gaussian16", well_c_out_str)
xyz6, xyz13 = automol.geom.coordinates(saddle_bc_geo, idxs=(6, 13))
disp = np.subtract(xyz6, xyz13)
xyz6 += 0.1 * disp
disp_geo = automol.geom.set_coordinates(saddle_bc_geo, {6: xyz6})

disp_geo_file = Path("disp_geom.xyz")
disp_geo_file.write_text(automol.geom.xyz_string(disp_geo))

automol.geom.display(saddle_bc_geo, label=True)
automol.geom.display(disp_geo, label=True)
automol.geom.display(well_c_geo, label=True)
```
